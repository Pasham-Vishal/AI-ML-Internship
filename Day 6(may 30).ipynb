```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
class Pie():
    sizes=[]
    df=pd.read_csv('Desktop/internship/notebooks/DataFiles/Match.csv')
    teams=np.unique(df.Team1)
    def wins(x,df):
        count=0
        for i in range(len(df)):
            if x == df.match_winner[i]:
                count=count+1
        return count
    for x in teams:
        ans=wins(x,df)
        sizes.append(ans)
    plt.pie(sizes,labels=teams,autopct='%1.2f%%',startangle=90)
    plt.show()
pie_plot=Pie
```
